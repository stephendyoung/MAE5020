```python
import numpy as np
import matplotlib.pyplot as plt

randomMat = np.random.randn(50,50)

U, S, VT = np.linalg.svd(randomMat, full_matrices=False)
    
plt.plot(S)
plt.xlabel("Index")
plt.ylabel("Singular Value")
plt.show()

means = []
medians = []
singularVals = []
U, S, VT = np.linalg.svd(np.random.randn(100,50,50))   






```

```python
plt.boxplot(S, showfliers=False)
plt.show()
#this will show 100 tests at each rank (1 to 50) and the distrubution of the singular value at each rank in the box and whisker plot
```

```python

i = 0
j = 0


means = [np.mean(S[:, i]) for i in range(S.shape[1])]
medians = [np.median(S[:, i]) for i in range(S.shape[1])]
        
fig, ax1 = plt.subplots()

line1, = ax1.plot(range(1, 51), means, color="blue", label="Mean")
ax1.set_xlabel('Rank', color='black')
ax1.set_ylabel('Mean', color='blue')  # Left Y-axis label

ax2 = ax1.twinx()
line2, = ax2.plot(range(1, 51), medians, color="orange", label="Median")
ax2.set_ylabel('Median', color='orange')  # Right Y-axis label

ax1.set_title("Mean and Median vs. Rank")

ax1.legend([line1, line2], ["Mean", "Median"], loc="upper right")

plt.show()
```
